```python
import pandas as pd
```

```python
df = pd.read_csv('D:\Semester 4\PASD\panic.csv')
df
```

```python
df.loc[df['Family History'] == 'Yes', 'Family History'] = 1
df.loc[df['Family History'] == 'No', 'Family History'] = 0
df.loc[df['Personal History'] == 'Yes', 'Personal History'] = 1
df.loc[df['Personal History'] == 'No', 'Personal History'] = 0
df.loc[df['Current Stressors'] == 'High', 'Current Stressors'] = 3
df.loc[df['Current Stressors'] == 'Moderate', 'Current Stressors'] = 2
df.loc[df['Current Stressors'] == 'Low', 'Current Stressors'] = 1
df.loc[df['Severity'] == 'Severe', 'Severity'] = 3
df.loc[df['Severity'] == 'Moderate', 'Severity'] = 2
df.loc[df['Severity'] == 'Mild', 'Severity'] = 1
df.loc[df['Impact on Life'] == 'Significant', 'Impact on Life'] = 3
df.loc[df['Impact on Life'] == 'Moderate', 'Impact on Life'] = 2
df.loc[df['Impact on Life'] == 'Mild', 'Impact on Life'] = 1
df.loc[df['Social Support'] == 'High', 'Social Support'] = 3
df.loc[df['Social Support'] == 'Moderate', 'Social Support'] = 2
df.loc[df['Social Support'] == 'Low', 'Social Support'] = 1

print(df)
```
